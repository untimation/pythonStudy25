```python
names = []
ko = []
en = []
ma = []
py = []
to = []
av = []
gr = []

run = True
name = ""
while run :
    prompt = """
    1. 성적추가
    2. 성적확인
    3. 성적수정
    4. 성적삭제
    5. 프로그램 종료

    Enter number (1~5) : """

    U = """
    1. 이름수정
    2. 국어수정
    3. 영어수정
    4. 수학수정
    5. 수정종료"""
    print(prompt)

    select = int(input("번호 입력 : "))
    if select == 1 :
        print("성적을 입력합니다.")
        print("이름을 입력해주세요.")
        name = input("이름 : ")
        names += [name]
        print("국어 성적을 입력해주세요.")
        kor = int(input("국어 : "))
        ko += [kor]
        print("영어 성적을 입력해주세요.")
        eng = int(input("영어 : "))
        en += [eng]
        print("수학 성적을 입력해주세요.")
        math = int(input("수학 : "))
        ma += [math]
        print("이름 : %s, 국어 : %d, 영어 : %d, 수학 : %d" % (name, kor, eng, math))
        tot = kor + eng + math
        to += [tot]
        avg = tot / 3
        av += [avg]
        if avg >= 90 :
            grade = "A"
            gr += [grade]
        elif avg >= 80 :
            grade = "B"
            gr += [grade]
        elif avg >= 70 :
            grade = "C"
            gr += [grade]
        elif avg >= 60 :
            grade = "D"
            gr += [grade]
        else :
            grade = "F"
            gr += [grade]
        print("성적이 등록되었습니다.")
    elif select == 2 :
        print("성적을 확인합니다.")
        print("이름을 확인합니다.")
        if name == "" :
            print("등록된 이름이 없습니다.")
        elif name == "%s" % name :
            print("이름 : %s, 국어 : %d, 영어 : %d, 수학 : %d" % (name, kor, eng, math))
            print("총점 : %d, 평균 : %.2f" % (tot, avg))
            print("성적을 확인하였습니다.")
        else :
            print("등록된 이름이 없습니다.")
    elif select == 3 :
        print("성적을 수정합니다.")
        print("이름을 확인합니다.")
        if name == "" :
            print("등록된 이름이 없습니다.")
        elif name == "%s" % name :
            print("확인 되었습니다.")
            print(U)
            Urun = True
            while Urun :
                U = int(input("번호를 입력해주세요 : "))
                if U == 1 :
                    print("이름을 수정합니다.")
                    name = input("이름 : ")
                    print("이름이 수정되었습니다.")
                elif U == 2 :
                    print("국어 성적을 수정합니다.")
                    kor = int(input("국어 : "))
                    print("국어 성적이 수정되었습니다.")
                elif U == 3 :
                    print("영어 성적을 수정합니다.")
                    eng = int(input("영어 : "))
                    print("영어 성적이 수정되었습니다.")
                elif U == 4 :
                    print("수학 성적을 수정합니다.")
                    math = int(input("수학 : "))
                    print("수학 성적이 수정되었습니다.")
                elif U == 5 :
                    print("수정을 종료합니다.")
                    Urun = False
                else :
                    print("옳바르지 않은 번호입니다. 다시 입력해 주세요.")
    elif select == 4 :
        print("성적을 삭제합니다.")
        print("이름을 확인합니다.")
        if name == "" :
            print("등록된 이름이 없습니다.")
        elif name == "%s" % name :
            name = ""
            kor = 0
            eng = 0
            math = 0
            print("삭제 되었습니다.")
        else :
            print("등록된 이름이 없습니다.")
    elif select == 5 :
        print("프로그램을 종료합니다.")
        run = False
    else :
        print("옳바르지 않은 번호입니다. \n다시 입력해 주세요.")
print(names, ko, en, ma, to, av, gr)
for name, kor, eng, math, tot, avg, grade in zip(names, ko, en, ma, to, av, gr):

    print("이름 : %s \n국어 : %d \n영어 : %d \n수학 : %d \n총합 : %d \n평균 : %.2f \n등급 : %s \n" % (name, kor, eng, math, tot, avg, grade))
```
