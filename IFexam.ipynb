```python
# if문은 제어문으로 참, 거짓을 판단하여 흐름을 조절함

# ex : 만약에 돈이 있으면 택시를 타고가고, 없으면 걸어가라!!!
#      if     True   print("택시를 탄다")  False   print("걸어감")

# if 문법 암기 :
# if 변수(참/거짓) :
# 탭(띄어쓰기) 실행문 -> 참일때 실행문
# 탭(띄어쓰기) 실행문 -> 참일때 실행문
# else : -> 거짓일때 실행문을 만듬
# 탭(띄어쓰기) 실행문 -> 거짓일때 실행문

money = False # 돈이 있다.
if money :   # 만약에 money 값안에 True이면 아래 들여쓰기 한 곳을 실행한다.
    print("택시를 타고 갑니다.") # True일때 출력문
else :       # 위에 if문에 False가 들어올때 아래 들여쓰기 한 곳 실행문
    print("걸어가라!!!")
print("학교에 도착하였습니다.")
```

```python
money = int(input("현재 얼마있나요?"))
#           input("프롬프트") : 키보드로 입력받은 값을 문자열로 보관
#       int() : 숫자로된 문자를 숫자화 -> 4000
# print("현재 주머니에 %d원이 있네요" % money)
if money >= 3000 :
    money -= 3000
    print("택시를 타고 갑니다.")
else :
    print("걸어갑니다.")
print("현재 남은돈")
print(money)
```

```python
card = True
trash = False
phone = False
money = int (input("현재 얼마있나요?"))
pocket = [trash, phone, money]

if money in pocket and money >= 5500:
    print("현금으로 택시를 타고간다.")
elif card in pocket:
    print("카드로 택시를 타고간다.")
elif money in pocket and money >= 1450:
    print("현금으로 버스를 타고간다.")
else:
    print("걸어간다.")
```
