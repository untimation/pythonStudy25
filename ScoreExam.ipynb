```python
# 성적처리용 프로그램을 만들어본다.
# 1. 키보드로 개인정보를 입력 받는다.
# 2. 키보드로 성적을 입력 받는다 (국어, 영어, 수학, 파이썬)
# 3. 입력된 정보를 확인한다. -> 잘못된 정보를 수정한다.
# 4. 통계를 만들어 출력한다.
# 5. 등급을 출력하고 종료한다.

print("=" * 50)
print("개인정보를 입력하세요!!!")
name = input("이름 : ")
age = input("나이 : ")
add = input("주소 : ")

print("=" * 50)
print("%s 님 환영합니다." % name)

print("=" * 50)
print("%s 님의 성적을 입력하세요 " % name)
kor = int(input("국어성적 : "))
eng = int(input("영어성적 : "))
mat = int(input("수학성적 : "))
pyt = int(input("파이썬성적 : "))

print("=" * 50)
print("성적을 확인합니다.")
print("국어 : %d \t 영어 : %d \t 수학 : %d \t 파이썬 : %d \t" % (kor, eng, mat, pyt))

print("=" * 50)
print("입력하신 성적을 수정하시겠습니까?")
modi = input("수정하시겠습니까? (y/n) : ")
if modi == "y":
    print("성적 수정을 진행합니다.")
    kor = int(input("국어성적 : "))
    eng = int(input("영어성적 : "))
    mat = int(input("수학성적 : "))
    pyt = int(input("파이썬성적 : "))
    print("성적이 수정 되었습니다. \n국어 : %d \t 영어 : %d \t 수학 : %d \t 파이썬 : %d \t" % (kor, eng, mat, pyt))

total = kor + eng + mat + pyt
avg = total / 4

print("=" * 50)
print("총점과 평균을 출력하시겠습니까?")
que = input("출력하시겠습니까? (y/n) : ")
if que == "y" :
    print("%s 님의 총점은 %d 평균은 %f 입니다. " % (name, total, avg))

print("=" * 50)
print("%s 님의 등급을 출력하시겠습니까?")
que1 = input("출력하시겠습니까? (y/n) :  ")
if que1 == "y" :
    if avg >= 90 : # 평균이 90점 이상이면 a등급
        grade = "A"
    elif avg >= 80 :
        grade = "B"
    elif avg >= 70 :
        grade = "C"
    elif avg >= 60 :
        grade = "D"
    else :
        grade = "F"

print("=" * 50)
print("%s 님의 등급은 %s 입니다. " % (name, grade))

print("=" * 50)
print("프로그램을 종료합니다.")
```
